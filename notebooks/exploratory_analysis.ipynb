# Electric Vehicle Population - Exploratory Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
try:
    df = pd.read_csv('../data/raw/Electric_Vehicle_Population_Data.csv')
    display(df.head())
except Exception as e:
    print(f'Error loading dataset: {e}')
```

```python
# Basic info
df.info()
```

```python
# Missing values
df.isnull().sum()
```

```python
# Distribution of EV makes
plt.figure(figsize=(12,6))
df['Make'].value_counts().head(10).plot(kind='bar', color='teal')
plt.title('Top 10 EV Makes')
plt.ylabel('Count')
plt.show()
```
